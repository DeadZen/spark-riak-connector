## Read CSV file

### FORMAT CONSOLIDATED CSV

#### Main table SDN.CSV

| Sequence | Column name | Description |
| -------- | ----------- | ----------- |
| 1        |  ent_num    | unique record identifier
| 2        |  SDN_Name   | name of entry
| 3        |  SDN_Type   | type of entry
| 4        |  Program    | sanctions program name
| 5        |  Title      | title of an individual
| 6        |  Call_Sign  | vessel call sign
| 7        |  Vess_type  | vessel type
| 8        |  Tonnage    | vessel tonnage
| 9        |  GRT        | gross registered tonnage
| 10       |  Vess_flag  | vessel flag
| 11       |  Vess_owner | vessel owner
| 12       |  Remarks    | remarks on entry

#### Address table ADD.CSV

|Sequence | Column name | Description
| ------- | ----------- | ------------
| 1       | Ent_num     | link to unique listing
| 2       | Add_num     | unique record identifier
| 3       | Address     | street address
| 4       | City		| city, state/province, zip
| 5       | Country     | country of address
| 6       | Add_remarks | remarks on address

#### Alternate identity table ALT.CSV

|Sequence | Column name | Description
| ------- | ----------- | ------------
| 1       | Ent_num     | link to unique listing
| 2       | alt_num     | unique record identifier
| 3       | alt_type    | type of alternate identity (aka, fka, nka)
| 4       | alt_name    | alternate identity name
| 5       | alt_remarks | remarks on alternate identity

- Record separator:              carriage return
- field (value) delimiter:       ,
- text value quotes:             "
- null:                          -0-

```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as mp
%matplotlib inline

sanctions_file = "/Users/quetzal/repos/ofac-analysis-demo/spark-riak-connector-demos-0.0.1-SNAPSHOT/sdn.csv"
address_file = "/Users/quetzal/repos/ofac-analysis-demo/spark-riak-connector-demos-0.0.1-SNAPSHOT/add.csv"
alt_file = "/Users/quetzal/repos/ofac-analysis-demo/spark-riak-connector-demos-0.0.1-SNAPSHOT/alt.csv"

sdn_names = ['id','SDN_Name','SDN_Type','Program','Title','Call_Sign','Vess_type','Tonnage','GRT','Vess_flag','Vess_owner','Remarks']
add_names = ['id','add_num','Address','City_State_ZIP','Country', 'Add_remarks']
alt_names = ['id','alt_num','Alt_type','Alt_name','Alt_remarks']

sdn = pd.read_csv(sanctions_file, na_values=['-0- '], names=sdn_names, skipfooter=1, engine='python', encoding='ASCII')
address = pd.read_csv(address_file, na_values=['-0- '], names=add_names, skipfooter=1, engine='python', encoding='ASCII')
alt = pd.read_csv(alt_file, na_values=['-0- '], names=alt_names, skipfooter=1, engine='python', encoding='ASCII')
[sdn.shape, address.shape, alt.shape]
```

```python
ofac = sdn.merge(address, how="left", on="id").merge(alt, how="left", on="id")
ofac.drop(['add_num', 'alt_num'], axis=1, inplace=True)
ofac.shape
```

## Create Riak bucket and push data into it

```python
from riak import RiakClient, RiakNode, RiakObject

rc = RiakClient(protocol='http', host='127.0.0.1', http_port=8098)
rc.ping()
```

```python
bucket = rc.bucket('OFAC-stats')
for ofacid in set(ofac['id']):
    obj = RiakObject(rc, bucket, str(ofacid))
    obj.content_type = 'text/json'
    obj.add_index('ofacid_int', ofacid)
    obj.data = ofac[ofac['id'] == ofacid].to_json()
    obj.store()
```

## Get data from Riak and create Spark RDD

```python
sjf = sc._gateway.jvm.com.basho.spark.connector.japi.SparkJavaUtil.javaFunctions(sc._jsc)
clz = sc._gateway.jvm.java.lang.Class.forName("java.lang.String")
rdd = sjf.riakBucket('OFAC-stats','default',clz).query2iRange("ofacid", 1, 20000)
```

```python
bucket = rc.bucket('OFAC-stats')
keys = set(ofac['id'])
ofac = pd.DataFrame()
for key in keys:
    ofac = ofac.append(pd.read_json(bucket.get(str(key)).data))
ofac.shape
```

```python
rdd = sc.parallelize(zip(set(ofac['id']), [ofac[ofac['id'] == ofacid].to_dict() for ofacid in set(ofac['id'])]))
rdd.take(1)
```

## Exploratory Statistics

### Let's find out number of banned individuals by each country 

```python
cntr = rdd.filter(lambda ent: notnan(ent[1]['Country'].values()[0]) and ent[1]['SDN_Type'].values()[0] == u'individual')
cntr = cntr.map(lambda ent: (ent[1]['Country'].values()[0], 1)).countByKey().items()
cntr = sorted(cntr, key=lambda x: -x[1])
```

```python
pos = np.arange(10)
lab = [country[0] for country in reversed(cntr[:10])]
num = [country[1] for country in reversed(cntr[:10])]

plt.figure(figsize=(8,6), dpi=100)
plt.barh(pos, num, align="center", color="skyblue")
plt.yticks(pos, lab)
plt.title("Top 10 countries by number of banned individuals")
for x,y in zip(num, pos):
    plt.text(x+4, y, x)
plt.show()
```

### How many vessels are in the OFAC list by country and vessel type?  

```python
vess = rdd.map(lambda ent: (ent[1]['Vess_type'].values()[0], ent[1]['Vess_flag'].values()[0]))
vess = vess.filter(lambda x: notnan(x[0]) and notnan(x[1]))
vess = vess.map(lambda ent: ((ent[0], ent[1].lower().title()), 1)).countByKey()
ves = list(set([v[0] for v in vess.keys()]))
countries = list(set([v[1] for v in vess.keys()]))
data = np.zeros((len(countries), len(ves)))
for v,c in vess.keys():
    data[countries.index(c)][ves.index(v)] = vess[(v,c)]
```

```python
plt.figure(figsize=(10,8), dpi=100)    
plt.pcolor(data, cmap=plt.cm.Reds,edgecolors='k')
plt.yticks(np.arange(len(countries))+0.5, countries)
plt.xticks(np.arange(len(ves))+0.5, ves, rotation=90)
plt.colorbar()
plt.title("Vessels in the OFAC list by country and vessel type")
plt.show()
```

### Plot a vessel tonnage histogram

```python
tonnage = rdd.map(lambda ent: ent[1]['Tonnage'].values()[0]).filter(lambda x: notnan(x))
bins, hist = tonnage.map(lambda t: float(t.replace(",", "")) if (isinstance(t, basestring)) else t).histogram(20)
```

```python
bins = np.array(bins)
hist = np.array(hist, dtype=np.dtype('Float64')) / sum(hist)
plt.figure(figsize=(16,8), dpi=100)
plt.bar((bins[:-1] + bins[1:]) / 2, hist, align='center', color="green", alpha=0.5, width=(bins[1] - bins[0])*0.9)
plt.title("Vessel tonnage histgram")
plt.show()
```
